## Постановка задачи

Дана функция $f(x) = ln(x) - cos(x)$ и начальный интервал изоляции корня $[0.8; 1.7]$. Требуется уточнить корни комбинированным и итерационным методами.

```python
import numpy as np

def f(x: float, a: float = 1) -> float:
    """Calculate f(x) = ln(x) - cos(x)"""
    return np.log(x) - a * np.cos(x)
```

## Проведение локализации корня и выбор правильного интервала изоляции

### Построение графика функции на интервале

```python
import matplotlib.pyplot as plt

def draw_graph() -> None:
    """Draw a graph for f(x) = ln(x) - cos(x)"""
    x_vals = np.linspace(0.8, 1.7)
    y_vals = f(x_vals)

    plt.figure(figsize=(7, 5))
    plt.axhline(0, color='black', linewidth=1, linestyle='-')
    plt.axvline(0, color='black', linewidth=1, linestyle='-')
    plt.plot(x_vals, y_vals, label=r'$f(x) = \ln x - \cos x$', color='blue')
    plt.xlim(0.5, 2.0)
    plt.ylim(-1.0, 1.0)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.2)
    plt.show()

draw_graph()
```

### Проверка достаточных условий для примения методов

Исследуем непрерывность функции:
- $ln(x)$ определена и непрерывна при $\textit{x}>0$,<br>
- $cos(x)$ определена и непрерывна при $\textit{x}∊R$,

$f(x)$ определена и непрерывна на $[0,8; 1,7]$.

Найдем первую и вторую производную:<br>
$f'(x)=\frac{1}{x}+sin(x)$ &emsp; $f''(x)=-\frac{1}{x^2}+cos(x)$.

Исследуем знак первой производной на интервале $[0,8; 1,7]$:

```python
def df(x: float, a: float = 1) -> float:
    """Calculate first derivative f(x) = 1/x + sin(x)"""
    return 1 / x + a * np.sin(x)

x1, x2 = df(0.8), df(1.7)
print(f"{x1:.5f}", f"{x2:.5f}")
```

$f'(0,8)\approx1,96736$ &emsp; $f'(1,7)\approx1,5799$

В обеих точках $f'(x)>0$ &rArr; $f(x)$ возрастает на интервале $[0,8; 1,7]$.

Исследуем знак второй производной на интервале $[0,8; 1,7]$:<br>

```python
def ddf(x: float, a: float = 1) -> float:
    """Calculate second derivative f(x) = 1/x^2 + cos(x)"""
    return -1 / x**2 + a * np.cos(x)

x1, x2 = ddf(0.8), ddf(1.7)
print(f"{x1:.5f}", f"{x2:.5f}")
```

$f''(0,8)\approx-0,86579$ &emsp; $f''(1,7)\approx-0,47487$

В обеих точках $f''(x)<0$ &rArr; $f(x)$ выпукла на интервале $[0,8; 1,7]$.

## Сужение/проверка интервала

Проверим свойства единственности корня на интервале изоляции:

```python
print(f"{(f(0.8)):.5f}, {(f(1.7)):.5f}")
print(f"{(df(0.8)):.5f}, {(df(1.7)):.5f}")
print(f"{(ddf(0.8)):.5f}, {(ddf(1.7)):.5f}")
```

Функция на левой граннице интервала $\approx-0.91985$, на правой границе $\approx0.65947$<br>
Первая производная на левой граннице интервала $\approx1.96736$, на правой границе $\approx1.57990$<br>
Вторая производная на левой граннице интервала $\approx-0.86579$, на правой границе $\approx-0.47487$

Функция на концах интервала имеет разные знаки, первая и вторая производные не меняют знак, следовательно на интервале изоляции существует единственный корень

Проверим выполнение неравенства $M\leq2m$:

```python
M = max(ddf(0.8), ddf(1.7))
m = min(df(0.8), df(1.7))
print(M <= m * 2)
```

Условие $M\leq2m$ выполняется: $1.57990<-0.86579*2$.
