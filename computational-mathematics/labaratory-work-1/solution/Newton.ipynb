```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Определяем функцию
def f(x):
    return np.log(x) - np.cos(x)

# Диапазон x (избегаем x <= 0)
x_vals = np.linspace(0.1, 10.0, 400)
y_vals = f(x_vals)

# Строим график
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label=r'$f(x) = \ln x - \cos x$', color='blue')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # ось X
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # ось Y (условно)

# Отметим найденный корень
root = calc_root(0.8)
plt.plot(root, 0, 'ro', label=f'Корень ≈ {root:.5f}')

plt.xlim(0.1, 3)
plt.ylim(-2, 2)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции $f(x) = \\ln x - \\cos x$')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import math

def f(x: float, a: float = 1) -> float:
    return math.log(x) - a * math.cos(x)

def df(x: float) -> float:
    return 1 / x + math.sin(x)

def newton(x: float) -> float:
    return x - f(x) / df(x)

def calc_root(x: float, eps: float = 1e-5) -> float:
    while True:
        if abs(newton(x) - x) < eps:
            return newton(x)
        x = newton(newton(x))

print(f"{calc_root(0.8):.6f}")
```
